```python
#problem 1

scores = [72 , 85 , 90 , 68 , 88]

print("Scores", scores )

average = sum(scores)/len(scores)
print("Average: ", average)

highest_score = max(scores)
lowest_score = min(scores)

print("Highest score: ", highest_score )
print("Lowest score: ", lowest_score )
```

```python
#problem 2

study_hours = [1 , 2 , 3 , 4 , 5]
exam_scores = [50 , 55 , 65 , 70 , 80]

for i in range (len(study_hours)):
  print("Input: ", study_hours[i], "hours -> Output: ", exam_scores[i], "marks")


```

```python
#problem 3

scores = [45 , 67 , 82 , 59]

def pass_fail(score):
  if score >= 60:
    return "Pass"
  else:
    return "Fail"

for score in scores:
  result = pass_fail(score)
  print("Input: ", score, "Output: ", result)
```

```python
#problem 4

sizes = [500 , 800 , 1000 , 1200]
prices = [50 , 80 , 100 , 130]

for i in range (len(sizes)):
  print("Size:", sizes[i], " Square meter -> Price:", prices[i],)

is_increasing = True
for i in range (1, len(prices)):
  if prices[i] <= prices[i-1]:
    is_increasing = False

if is_increasing:
  print("Positive trend")
else:
  print("Negative trend")


```

```python
#problem 5

study_hours = [2 , 4 , 6 , 8]
scores = [55 , 65 , 75 , 85]

for i in range (len(study_hours)):
  print("Increased score for every two hours: ", scores[i]+1 - scores[i])


```

```python
scores = [35 , 58 , 62 , 74 , 89 , 45]

for i in range (len(scores)):
    if scores[i] >= 70:
        print("High")
    elif scores[i] >= 50:
         print("Medium")
    else:
        print("Low")
```
